## Whistler Blackcomb resort status

This will be used as an example of how to grab real-time data from https://whistlerblackcomb.com and display it here.

```python
#imports needed for scrapping web pages
import requests, bs4
```

```python
#get the website page

#this page has the mountain conditions for whistler blackcomb
res = requests.get("https://www.whistlerblackcomb.com/the-mountain/mountain-conditions/snow-and-weather-report.aspx")
```

```python
#use Beautiful Soup to convert this into a soup
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
soup
```

```python
def get_data_subclass(data,data_string):
    #print(datapoint.find(class_=data_string).text)
    return data.find(class_=data_string).text



#select our live temperature values now




#go over the different weather stations displayed
for datapoint in soup.select('.live_temperature__station_container'):

    #find the name of the station
    station_string = datapoint.find(class_='live_temperature__station_name')
    #print(station_string.text.lstrip())
    station_name = station_string.text.lstrip().rstrip()
    
    #get the live temperature
    soup_string = datapoint.find(class_='live_temperature__international_only')
    #print(soup_string.text)
    #my_list = soup_string.text.split(" ")
    my_list = get_data_subclass(datapoint,'live_temperature__international_only').split(" ")

    temp = my_list[0]
    units = my_list[1]
    #print(my_list[0])
    #print(my_list[1])

    #figure out when it was last updated.
    updated_string = datapoint.find(class_='live_temperature__last_updated')
    updated = updated_string.text.lstrip().rstrip()

    #wind_string = datapoint.find(class_='icon-weather-windy')
    #wind = wind_string.text.lstrip().rstrip()

    print(f"The weather station {station_name} has temperature of {temp} with units {units}. {updated}")

    
```

```python
#more imports needed. Whistler Peak uses Javascript to load the temperatures, so we can't just use requests.get to 
#obtain the webpage. It won't update with the temperature values
import selenium as se
from selenium import webdriver

```

```python
#because PhantomJS has been deprecated from 
#now we use the suggestions in https://stackoverflow.com/questions/48537028/selenium-how-to-use-headless-chrome-on-aws
#and https://stackoverflow.com/questions/48537028/selenium-how-to-use-headless-chrome-on-aws

#need a headless firefox browser to properly render things
def createHeadlessFirefoxBrowser():
     options = webdriver.FirefoxOptions()
     options.add_argument('--headless')
     return webdriver.Firefox(options=options)

browser = createHeadlessFirefoxBrowser()

browser.get("https://whistlerpeak.com/temps/")
html = browser.page_source
#soup = bs4.BeautifulSoup(html, 'lxml')
soup = bs4.BeautifulSoup(html, features="html.parser")
#a = soup.find('section', 'wrapper')
```
